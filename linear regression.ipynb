```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/housing.csv')
value = df.iloc[:500 , :6]
label = df.iloc[:500 , 7:8]
```

```python
value_train, value_test, label_train, label_test = train_test_split(value, label, test_size=0.2, random_state=42)
lr = LinearRegression()
lr.fit(value_train, label_train)
```

```python
label_pred = lr.predict(value_test)
mse = mean_squared_error(label_test, label_pred)
print(f"Mean Squared Error: {mse}")
plt.scatter(label_test, label_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Prices")
plt.show
r2 = r2_score(label_test, label_pred)
print(f"R-squared: {r2}")
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error: {rmse}")
```
